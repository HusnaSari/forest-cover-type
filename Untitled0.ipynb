5. Aspen (Titrek Kavak)
6. Douglas-fir (Douglas Köknarı)
7. Krummholz (Alp orman sınırındaki kıvrık bodur ağaçlar)

**Elevation:**  Metre cinsinden yükseklik.(Ormanın yer aldığı bölgenin )

**Aspect:** Bir eğimin veya yüzeyin baktığı yönü belirtir. Bu yön, bir pusula üzerindeki dereceyle ifade edilir (0° ile 360° arasında).

0°: Kuzeyi gösterir.
90°: Doğuyu gösterir.
180°: Güneyi gösterir.
270°: Batıyı gösterir.
arazinin hangi yöne doğru eğimli olduğunu ölçer ve "azimut yönü" olarak adlandırılır

**Slope:** Eğim oranı, yer yüzeyinin eğim açısını belirtir. Arazi düz mü yoksa daha dik?

**Horizontal_Distance_To_Hydrology:** En yakın yüzey suyu kaynağına yatay mesafe

**Vertical_Distance_To_Hydrology:** En yakın yüzey suyu kaynağına dikey mesafe

**Horizontal_Distance_To_Roadways:** En yakın yola yatay mesafe

**Hillshade_9am:** Sabah 9'daki gölge indeksi (0 ile 255 arasında), yaz gündönümü (Yılın en uzun gündüzü)

**Hillshade_Noon:** Öğlen 12'deki gölge indeksi (0 ile 255 arasında), yaz gündönümü.

**Hillshade_3pm:** Öğleden sonra 3'teki gölge indeksi (0 ile 255 arasında), yaz gündönümü.

**Horizontal_Distance_To_Fire_Points:** En yakın yangın başlama noktalarına olan yatay mesafe (metre cinsinden).

**Wilderness_Area:** Vahşi yaşam alanı (4 tane sütun).

**Soil_Type:**  Toprak Türü (40 tane sütun )

```python
numeric_features =[feature for feature in veri_seti.columns if veri_seti[feature].dtype!='O']
categorical_features =[feature for feature in veri_seti.columns if veri_seti[feature].dtype=='O']

print('{} adet sayısal nitelik var : {}'.format(len(numeric_features),numeric_features))
print('{} adet kategorik nitelik var : {}'.format(len(categorical_features),categorical_features))
```

**Tek Değişkenli Analiz (Univariate Analysis) :** Tek bir değişkenin değerlerinin dağılımını anlamak

```python
plt.figure(figsize=(30,20))
plt.suptitle('Sayısal Niteliklerin Tek Değişkenli Analizi',fontsize=20, fontweight='bold', alpha=0.8, y=1)

for i in range(0, len(numeric_features)):
  plt.subplot(5,3,i+1)
  sns.kdeplot(x=veri_seti[numeric_features[i]], color='purple')
  plt.xlabel(numeric_features[i])
  plt.tight_layout()
```

```python
plt.figure(figsize=(30,15))
plt.suptitle('Katagorik Niteliklerin Tek Değişkenli Analizi',fontsize=20, fontweight='bold', alpha=0.8, y=1)

for i in range(0, len(categorical_features)):
  plt.subplot(2,3,i+1)
  sns.countplot(x=veri_seti[categorical_features[i]], palette = 'pastel')
  plt.xlabel(categorical_features[i])
  plt.tight_layout()
```

```python
# Discrete (Ayrık) özelliklerin tanımlanması
discrete_features = [feature for feature in numeric_features if len(veri_seti[feature].unique()) <= 25 and len(veri_seti[feature].unique()) > 3]

# Continuous (Sürekli) özelliklerin tanımlanması
continuous_features = [feature for feature in numeric_features if len(veri_seti[feature].unique()) > 25]

# Encoded Categorical (Kodlanmış Kategorik) özelliklerin tanımlanması
encoded_categorical = [feature for feature in numeric_features if len(veri_seti[feature].unique()) <= 3]

# Özelliklerin sayısını ve isimlerini yazdırma
print('{} adet ayrık nitelik var : {}'.format(len(discrete_features), discrete_features))
print('\n{} adet sürekli nitelik var : {}'.format(len(continuous_features), continuous_features))
print('\n{} adet kodlanmış kategorik nitelik var : {}'.format(len(encoded_categorical), encoded_categorical))
```

```python
from scipy.stats import chi2_contingency

ki_kare_testi= []
for feature in categorical_features:
    if chi2_contingency(pd.crosstab(veri_seti['Cover_Type'], veri_seti[feature]))[1] < 0.05: #4 değer dönüyor p değeri ikinci yani 1. indexteki
        ki_kare_testi.append('Bağımsızlık Hipotezi Reddedildi (korelasyon var).')
    else:
        ki_kare_testi.append('Bağımsızlık hipotezi reddedilemedi (korelasyon yok).')
result = pd.DataFrame(data=[categorical_features, ki_kare_testi]).T
result.columns = ['Column', 'Hipotez Sonucu']
result
```

**Çıkarımlar:**

**Slope( Eğim) :** tüm kutu grafiklerinden en dar aralığa sahip olanıdır. Bu, medyanın ve ortalamanın birbirine çok yakın olduğu anlamına gelir.

**Yola Olan Yatay Mesafe**: Tüm özellikler arasında en yaygın veri dağılımına sahiptir. Bunun nedeni, en yüksek standart sapma değerine sahip olmasıdır. Yani, bu özellik diğer özelliklere göre daha fazla değişkenlik gösterir.

**Yangın Noktalarına Olan Yatay Mesafe:** Yola mesefaye benzer bir görünüme sahiptir, ancak maksimum değere sahiptir. Yani, yangın noktalarına olan mesafeler genel olarak yola olan mesafelerden daha kısadır, ancak bazı noktalar çok daha uzak olabilir.


Yola Olan Yatay Mesafenin son %50'si, Yangın Noktalarına Olan Yatay Mesafeye göre çok daha yaygın ve daha az yoğun olduğu için daha yüksek bir standart sapma değerine sahiptir.






**Keman Grafiği:**

1. Sayısal bir değişkenin dağılımının grafiksel gösterimidir.

2. Keman grafiği, verilerin dağılımının şeklini, verilerin ortasını, verilerin yayılımını ve verilerin simetrik veya çarpık olup olmadığını gösterir.

```python
num_features = veri_seti.iloc[:,:10]
sns.set_style("darkgrid", {'grid.color': '.1'})

target = veri_seti['Cover_Type']
features = num_features.columns

for i in range(0, len(features)):
  plt.subplots(figsize=(10,9))
  sns.violinplot(data=num_features, x=target, y=features[i],hue=target)
  plt.xticks(size=10)
  plt.yticks(size=10)
  plt.xlabel('Orman Örtüsü Türleri', size=18)
  plt.ylabel(features[i], size=18)
  plt.show()
```

**Kavak/Söğüt** orman örtüsü, 2000m-2500m yükseklikler arasında en fazla örtüye sahiptir.

**Krummholz** En yükseklerde gözlemlenen orman örtüsü ve bu gözlemler -2800m kadar düşük ve -3800m kadar yüksek rakımlarda gerçekleşmiştir.

**Aspect(Bakı)** Her sınıf için normal bir dağılıma sahip

**Slope(Eğim)** Tüm özellikler içinde en düşük maksimum değere sahiptir.
Tüm sınıflar 0-20 derece arasında yoğun eğim gözlemlerine sahiptir

**Gölgelenmeler** Negatif çarpık bir dağılıma sahip ve her sınıf için gözlemlerin büyük çoğunluğunda 200-250 indeks değeri arasında maksimum değere ulaşıyorlar

```python
fig, axs = plt.subplots(ncols=2, figsize=(15, 4))
fig.suptitle("Positive Correlations", fontsize=20)
sns.lineplot(x="Aspect", y="Hillshade_3pm", data=veri_seti, color='green', ax=axs[0])
sns.lineplot(x="Hillshade_Noon", y="Hillshade_3pm", data=veri_seti, color='green', ax=axs[1])
```
